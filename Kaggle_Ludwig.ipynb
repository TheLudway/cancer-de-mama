- ¿Cuál es la distribución de los tamaños de los tumores (Tumor Size) en el conjunto de datos?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Kaggle/Breast_Cancer.csv', delimiter=',', encoding='unicode_escape')
df
```

```python
df.columns
```

```python

plt.scatter(df['Tumor Size'], df['Survival Months'])
plt.xlabel('Tamaño del tumor')
plt.ylabel('Meses de esperanza de vida')
plt.title('Gráfico de dispersión entre la edad y el tamaño del turmo')
plt.show()
```

```python
plt.scatter(df['Survival Months'], df['Tumor Size'])
plt.xlabel('Meses de supervivencia')
plt.ylabel('Tamaño del tumor')
plt.title('Gráfico de dispersión entre los meses de supervivencia y el tamaño del tumor')
plt.show()
```
