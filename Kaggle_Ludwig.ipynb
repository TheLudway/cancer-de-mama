- ¿Cuál es la distribución de los tamaños de los tumores (Tumor Size) en el conjunto de datos?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
df = pd.read_csv('Kaggle/Breast_Cancer.csv', delimiter=',', encoding='unicode_escape')
df
```

```python
df.columns
```

## Relación entre la esperanza de vida (meses) y el tamaño del tumor (milímetros)

```python
plt.scatter(df['Tumor Size'], df['Survival Months'])
plt.xlabel('Tamaño del tumor (mm)')
plt.ylabel('Meses de esperanza de vida')
plt.title('Gráfico de dispersión entre la esperanza de vida y el tamaño del tumor')
plt.show()
```

# Teorema central del límite

```python
promedio = df['Tumor Size'].mean()
std_dev = df['Tumor Size'].std()

num_samples = 1000
sample_means = [np.mean(np.random.choice(df['Tumor Size'], size=len(df['Tumor Size']))) for _ in range(num_samples)]

plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, promedio, std_dev / np.sqrt(len(df['Tumor Size'])))
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Media de la muestra')
plt.ylabel('Densidad')
plt.title('Aproximación de la distribución a una normal')

plt.show()
```
