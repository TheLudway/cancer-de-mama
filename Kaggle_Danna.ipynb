#INTERPRETACION DE LA MEDIANA 
Las muejeres de estamuestra tienen 54 años o menos 

```python
# Vamos a separar las edades pr rango de 10 en 10 y vamos a contar cauntas mujeres estan dentro de dichos rangos 
r1 = df.loc[:,["Age", "Status"]]

# Primer rango 30 - 40 años
r1v  = r1[(r1["Age"]< 40) & (r1["Status"] == "Alive")].count(axis = 0)
r1v = r1v["Age"]
r1m = r1[(r1["Age"]< 40) & (r1["Status"] == "Dead")].count(axis = 0)
r1m = r1m["Age"]

# Segundo rango 40 - 50 años
r2v  = r1[(r1["Age"]< 50) & (r1["Age"]>=40) & (r1["Status"] == "Alive")].count(axis = 0)
r2v = r2v["Age"]

r2m = r1[(r1["Age"]< 50) & (r1["Age"]>=40) & (r1["Status"] == "Dead")].count(axis = 0)
r2m = r2m["Age"]

# Tercer rango 50 - 60 años
r3v  = r1[(r1["Age"]< 60) & (r1["Age"]>=50) & (r1["Status"] == "Alive")].count(axis = 0)
r3v = r3v["Age"]

r3m = r1[(r1["Age"]< 60) & (r1["Age"]>=50) & (r1["Status"] == "Dead")].count(axis = 0)
r3m = r3m["Age"]

# Cuarto rango 60 - 70 años
r4v  = r1[(r1["Age"]< 70) & (r1["Age"]>=60) & (r1["Status"] == "Alive")].count(axis = 0)
r4v = r4v["Age"]

r4m = r1[(r1["Age"]< 70) & (r1["Age"]>=60) & (r1["Status"] == "Dead")].count(axis = 0)
r4m = r4m["Age"]

print(r2v)
```

```python
# Hallamos el porcentaje de mujeres que viven o mueren dentro de los rangos de edad anteriores 
rangos = {
    "Rangos" : ["De 30-40 años","De 40-50 años","De 50-60 años","De 60-70 años"],
    "Vivos" : [r1v,r2v,r3v,r4v],
    "%vivos" : [(r1v/(r1v+r1m))*100,(r2v/(r2v+r2m))*100,(r3v/(r3v+r3m))*100,(r4v/(r4v+r4m))*100],
    "Muertos" : [r1m,r2m,r3m,r4m],
    "%muertos" : [(r1m/(r1v+r1m))*100,(r2m/(r2v+r2m))*100,(r3m/(r3v+r3m))*100,(r4m/(r4v+r4m))*100]
}

dr = pd.DataFrame(rangos)
dr
```

#INTERPRETACION DE LOS PROMEDIOS 
Dentro de los rangos de edad se evidencia que quienes presentan una mayor esperanza de vida son las mujeres con edades entre los 40 a 60 años de edad ya que son los rangos de edad que presentan un menor porcentaje de mortalidad y un mayor porcentaje de supervivencia

```python
import math

arry = df['Age']
print(df["Age"])


def calculate_percentile(arry, percentile):
    size = len(arry)
    return sorted(arry)[int(math.ceil((size * percentile) / 100)) - 1]


percentile_25 = calculate_percentile(arry, 25)
percentile_50 = calculate_percentile(arry, 50)
percentile_75 = calculate_percentile(arry, 75)
print("The 25th percentile is:", percentile_25)
print("The 50th percentile is:", percentile_50)
print("The 75th percentile is:", percentile_75)

```
