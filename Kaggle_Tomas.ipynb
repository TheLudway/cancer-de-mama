```python
muestras = []
for _ in range(1000):
    muestras.append(data['Age'].sample(30).mean())
```

```python
muestras_df = pd.DataFrame(muestras, columns=['Muestra'])
```

```python
sns.histplot(muestras_df['Muestra'], color="#FFC0CB")
plt.title('Distribución de las medias muestrales (Teorema Central del Límite)')
plt.show()
pd.option_context('mode.use_NaN_as_na', True)
```

```python
muestras_df = pd.DataFrame(muestras, columns=['Muestra'])
```

```python
sns.ecdfplot(data=muestras_df, color="#FFC0CB")
plt.title('Distribución de las medias muestrales (Teorema Central del Límite)')
plt.show()
pd.option_context('mode.use_inf_as_na', True)
```
